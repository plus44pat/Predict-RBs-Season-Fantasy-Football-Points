```python
mport pandas as pd

# Assuming your dataframe is named week_on_week_data

# Step 1: Filter data for seasons 2013-2022 and position labeled as RB. Now this is where I change it to be from 2013-2022, which is 10 seasons
filtered_data = week_on_week_data[(week_on_week_data['season'].between(2013, 2022)) & (week_on_week_data['position'] == 'RB')]


'''
Aggregate based on season and player_name. This is me aggregating all the data that I would already know going into
the year. The only exception is fantasy points, which is my target variable so I want to aggregate it now and not look
at past data.  Also, there is some data I already looked at in the past regarding the pro bowl.
'''
aggregated_data = filtered_data.groupby(['season', 'player_name']).agg({
    'player_id': 'first',
    'position': 'first',
    'team': 'first',
    'full_name': 'first',
    'fantasy_points': 'sum',
    'fantasy_points_ppr': 'sum',
    'head_coach': 'first',
    'first_name': 'first',
    'last_name': 'first',
    'birth_date': 'first',
    'height': 'first',
    'weight': 'first',
    'years_exp': 'first',
    'game_type': 'first',
    'entry_year': 'first',
    'rookie_year': 'first',
    'draft_number': 'first',
    'Past Year': 'first',  # Was the running back a pro bowler last year
    'Past 3 Years': 'first',  #How many times was the running back a pro bowler in the previous three years
    'round': 'first',
    'college': 'first',
    'College Conference': 'first',
    'Offense_Past_Year_Sum': ['sum', 'median', 'mean'],  #How many teammates to the RB on offense is a Pro Bowler
    'Offense_Past_3_Years_Sum': ['sum', 'median', 'mean'],  #How many cumulative pro bowls does the current offense have in last 3 years
    'Defense_Past_Year_Sum': ['sum', 'median', 'mean'],  # For the defenses against the RB, how many Pro Bowls do they have?
    'Defense_Past_3_Years_Sum': ['sum', 'median', 'mean'], # For the defenses against the RB, what is the total amount of Pro Bowls in the last 3 years
    'Defensive_Line_Past_3_Years_Sum': ['sum', 'median', 'mean'],#Same concept as Defense, just being more specific to defensive line
    'Defensive_Line_Past_Year_Sum': ['sum', 'median', 'mean'], #Same concept as Defense, just being more specific to defensive line
    'Offensive_Line_Past_Year_Sum': ['sum', 'median', 'mean'],  #Same concept as offense but being more specific to O-Line
    'Offensive_Line_Past_3_Years_Sum': ['sum', 'median', 'mean']  #Same concept as offense but being more specific to O-Line

    '''
    The reason I want to be specific for the Defensive Line and the Offensive Line is because that is how the Running back will
    create holes and gain more yards or it is how they will get no yards because the defensive line is really good
    '''
})

# Step 2: Reset index to include 'season' and 'player_name' as columns
aggregated_data.reset_index(inplace=True)

# Clean column names
aggregated_data.columns = ['_'.join(col).strip() for col in aggregated_data.columns.values]

# Remove '_first' from column names
aggregated_data.columns = [col.replace('_first', '') for col in aggregated_data.columns]

# Clean up aggregation suffixes
aggregated_data.columns = [col.replace('_sum_sum', '_sum').replace('_sum_median', '_median').replace('_sum_mean', '_mean') for col in aggregated_data.columns]

# Rename fantasy_points and fantasy_points_ppr columns
aggregated_data.rename(columns={'fantasy_points_sum': 'current_fantasy_points',
                                'fantasy_points_ppr_sum': 'current_fantasy_points_ppr',
                               'season_': 'season',
                               'player_name_': 'player_name'}, inplace=True)


# Print the cleaned DataFrame
print(aggregated_data.info())
```
